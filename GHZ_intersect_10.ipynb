```python
from qiskit_ionq import IonQProvider, ionq_job
ionq_provider = IonQProvider("mycsEhnoIgARQdch53Zm7KJUePwFD0Wr")
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit.tools.visualization import plot_histogram, array_to_latex
from qiskit.visualization import plot_state_city, plot_bloch_multivector
from qiskit.quantum_info import Statevector, DensityMatrix, Operator
from matplotlib import pyplot as plt
import numpy as np
import time
```

```python
qr = QuantumRegister(5,'q')
cr = ClassicalRegister(5,'c')
qc = QuantumCircuit(qr,cr)

qc.h(0)
qc.cx(0,1)
qc.cx(0,2)
qc.cx(0,3)
qc.cx(0,4)

qc.measure_all(add_bits=False)
qc.draw(output='mpl')
```

```python
qr = QuantumRegister(5,'q')
cr = ClassicalRegister(5,'c')
qcc = QuantumCircuit(qr,cr)

qcc.h(4)
qcc.cx(4,3)
qcc.cx(4,2)
qcc.cx(4,1)
qcc.cx(4,0)

qcc.measure_all(add_bits=False)
qcc.draw(output='mpl')
```

```python
sim_backend = ionq_provider.get_backend('ionq_simulator')

qpu_backend = ionq_provider.get_backend('ionq_qpu')
```

```python
ID = []
```

```python
for i in range(10):
    job1 = qpu_backend.run(qc,shots = 1024)
    job2 = qpu_backend.run(qcc,shots = 1024)
    ID.append(job1.job_id())
    ID.append(job2.job_id())
```
